# Circular mean
Since the arithmetic mean is not always appropriate for angles, the following method can be used to obtain both a mean value and measure for the variance of the angles.
This method is valid for angles in the range (-180, 180] degrees or (-pi, pi] radians.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def circular_mean(angles):
    sn_sum = np.sum(np.sin(angles))
    cs_sum = np.sum(np.cos(angles))
    avg = np.arctan2(sn_sum, cs_sum)
    return avg
```

```python
N = 2
angles = np.random.random(size=N) * 2 * np.pi - np.pi
avg_angle = circular_mean(angles)
print("Angles = ", angles)
print("Avg = ", avg_angle)

fig, ax = plt.subplots()
circle = plt.Circle((0, 0), 1, fill = False )
ax.set_aspect(1)
ax.add_artist(circle)

for i in range(0, N):
    plt.arrow(x=0, y=0, dx=np.cos(angles[i]), dy=np.sin(angles[i]), width=.05, length_includes_head=True) 
plt.arrow(x=0, y=0, dx=np.cos(avg_angle), dy=np.sin(avg_angle), width=.05, length_includes_head=True, color='red') 

plt.xlim( -1.2 , 1.2 )
plt.ylim( -1.2 , 1.2 )
plt.show()
```
